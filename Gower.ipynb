Gower distance is a metric used to calculate the similarity between two data points when those points contain a mix of data types (e.g., numbers, categories, and yes/no values).

Most distance formulas like Euclidean or Cosine only work if everything is a number. If your dataset looks like a spreadsheet with names, ages, and "Yes/No" columns, those formulas will break. Gower distance is the "Swiss Army Knife" designed specifically for that scenario.

```python
import pandas as pd
import gower

# Create a small dataset of customer orders
# Row 0 and Row 1 are very similar, Row 2 is different
data = pd.DataFrame({
    'total_price': [1200, 1250, 5000],
    'location': ['Colombo', 'Colombo', 'Kandy'],
    'is_priority': [True, True, False]
})

# Calculate the distance matrix
# This compares every row against every other row
distance_matrix = gower.gower_matrix(data)

print("Gower Distance Matrix:")
print(distance_matrix)

# To see the distance between the first and second order:
dist_0_1 = distance_matrix[0, 1]
print(f"\nDistance between Order 0 and Order 1: {dist_0_1:.4f}")
```
